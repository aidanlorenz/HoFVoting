```python
import numpy as np
import pandas as pd  
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
all_data = pd.read_csv('all_data.csv')
```

```python
all_data.head()
```

Remove all players whose primary position was pitcher (indicated by *1 in the position summary column).

```python
all_data_clean = all_data[~all_data['Pos\xa0Summary'].str.contains('*1', regex = False)]
```

```python
all_data_clean.head()
```

```python
all_data_clean.columns
```

Now we want to drop all the columns that have to do with pitching stats.

```python
all_data_clean = all_data_clean.drop(['W','L','ERA','ERA+','WHIP','G.1','GS','SV','IP','H.1','HR.1','BB.1','SO'], axis = 1)
```

```python
all_data_clean.head()
```

```python
all_data_clean.columns
```

Now we want to drop the irrelevant columns Rk, Name, Pos Summary (it would be interesting to try to incorporate position but we'll leave that for another time as the Pos Summary columns would require some cleaning), and Votes (as we really care about %vote).

```python
all_data_clean = all_data_clean.drop(['Rk','Name','Pos\xa0Summary', 'Votes'], axis = 1)
```

```python
all_data_clean
```

Replace %vote with a decimal, and YoB ordinal with an integer.

```python
all_data_clean['%vote'] = all_data_clean['%vote'].str.replace('%', '').astype(float) / 100
all_data_clean['YoB'] = all_data_clean['YoB'].str.replace(r'\D', '', regex=True).astype(int)
```

```python
all_data_clean
```

And export our results.

```python
all_data_clean.to_csv('all_data_clean.csv', index = False)
```
